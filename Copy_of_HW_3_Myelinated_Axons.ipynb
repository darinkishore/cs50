```python
# Import packages for running Hodgkin-Huxley simulations (NEURON),
# plotting (matplotlib, mpld3), and neatly displaying python
# dictionaries (pprint)

%reset
!pip install neuron
!pip install mpld3
import pprint # "pretty-print"


import matplotlib.pyplot as plt, mpld3
import pprint # "pretty-print"
from neuron.units import ms, mV

# The Python interface to NEURON goes through the “h” module.
from neuron import h
h.load_file('stdrun.hoc')
```

```python

# Define the physical properties of our neuron
soma = h.Section(name='soma')   # Use NEURON h module to define a soma (cell body)
soma.nseg = 1               # "nseg" = number of segments/compartments - here, the soma consists of one compartment
soma.diam = soma.L = 20     # length and diameter of (cylindrical) soma compartment, in microns

axon_myelin_1 = h.Section(name='axon')      # Use NEURON h module to define a axon segments
axon_myelin_2 = h.Section(name='axon')
axon_myelin_3 = h.Section(name='axon')
axon_no_myelin_1 = h.Section(name='axon')
axon_no_myelin_2 = h.Section(name='axon')
axon_no_myelin_3 = h.Section(name='axon')

# We'll define all axon compartments as having 10 segments
axon_myelin_1.nseg = 10;axon_myelin_2.nseg = 10;axon_myelin_3.nseg = 10;
axon_no_myelin_1.nseg = 10;axon_no_myelin_2.nseg = 10;axon_no_myelin_3.nseg = 10;

# Inserting Hodgkin-Huxley channels in nodes, but not internodes, and concatenating axon segments
for sec in [soma, axon_no_myelin_1, axon_no_myelin_2, axon_no_myelin_3]:
    sec.insert('hh')            # This command inserts 'hh' channels: Hodgkin-Huxley sodium, potassium, and leakage channels in the specified segments

# Inserting passive channels in the myelinated internodes
for sec in [axon_myelin_1, axon_myelin_2, axon_myelin_3]:
    sec.insert('pas')

# Axon diameter is a big bigger in at nodes (no myelin)
diameter_myelin=2;
diameter_no_myelin=1;

axon_myelin_1.diam=diameter_myelin
axon_myelin_2.diam=diameter_myelin
axon_myelin_3.diam=diameter_myelin

axon_no_myelin_1.diam=diameter_no_myelin
axon_no_myelin_2.diam=diameter_no_myelin
axon_no_myelin_3.diam=diameter_no_myelin


# Myelinated parts of axon (internodes) are longer than unmyelinated parts (nodes)
#
# Note that these values, and others in this simulation, are not realistic
# (the length of nodes and internodes are shorter than this); I've
# chosen these to illustrate paticular aspects of myeliated axons
myelin_lengt=2000;no_myelin_length=100;
axon_myelin_1.L=myelin_lengt;axon_myelin_2.L=myelin_lengt;axon_myelin_3.L=myelin_lengt;
axon_no_myelin_1.L=no_myelin_length;axon_no_myelin_2.L=no_myelin_length;axon_no_myelin_3.L=no_myelin_length;

# Myelinated parts of axon (internodes) have smaller membrane capacitance than unmyelinated parts (nodes)
myelin_cm=.2;no_myelin_cm=1;

axon_myelin_1.cm=myelin_cm
axon_myelin_2.cm=myelin_cm
axon_myelin_3.cm=myelin_cm
axon_no_myelin_1.cm=no_myelin_cm
axon_no_myelin_2.cm=no_myelin_cm
axon_no_myelin_3.cm=no_myelin_cm

# Now we have defined single segments of myelinated and unmyelinated axon.
# Here we'll make copies of each of these so that our neuron will have two myelinated
# segments and two unmyelinated:
#
# S-U1-M1-U2-M2-U3-M3 (where S=soma, U=unmyelinated)
#
axon_no_myelin_1.connect(soma(1))             # connect first UNmyelinated axon segment to the soma
axon_myelin_1.connect(axon_no_myelin_1(1))    # connect first myelinated segment to end of the first unmyelinated segment
axon_no_myelin_2.connect(axon_myelin_1(1))    # connect second unmyelinated segment to end of the first myelinated segment
axon_myelin_2.connect(axon_no_myelin_2(1))    # connect second myelinated segment to end of the second unmyelinated segment
axon_no_myelin_3.connect(axon_myelin_2(1))    # connect second unmyelinated segment to end of the first myelinated segment
axon_myelin_3.connect(axon_no_myelin_3(1))    # connect second myelinated segment to end of the second unmyelinated segment

# Draw a (crude) diagram of the model neuron
h.topology()

stim = h.IClamp(soma(0.5))  # Use NEURON h module to insert a stimulating electrode ("stim") into the soma compartment to perform current-clamp ("IClamp") experiments

# Below, we set parameters of "stim" to define the timing and amplitude of injected current
stim.delay = 10 * ms        # Delay until onset of stimulation in msec
stim.dur = 1 * ms           # Duration of stimulation in msec
stim.amp = 5                # Amplitude of stimulation in nanoAmps

```

```python
# Define which parameters we want to record
t = h.Vector().record(h._ref_t)           # Record time (t), in msec
v0 = h.Vector().record(soma(0)._ref_v)    # Record voltage at the soma
v1 = h.Vector().record(axon_no_myelin_1(0.5)._ref_v)  # Record voltage at the in the middle (0.5) of the first myelinated segment
v2 = h.Vector().record(axon_myelin_1(0.5)._ref_v)  # Record voltage at the in the middle (0.5) of the first unmyelinated segment
v3 = h.Vector().record(axon_no_myelin_2(0.5)._ref_v)  # Record voltage at the in the middle (0.5) of the SECOND myelinated segment
v4 = h.Vector().record(axon_myelin_2(0.5)._ref_v)
v5 = h.Vector().record(axon_no_myelin_3(0.5)._ref_v)
v6 = h.Vector().record(axon_myelin_3(0.5)._ref_v)

# Initialize membrane voltage and run the simulation
h.finitialize(-65 * mV)   # initial membrane voltage
h.continuerun(100 * ms)   # duration of simulation

# Plot data below
plt.plot(t, v0, label='soma.v')
plt.plot(t, v1, label='axon_no_myelin_1(0.5).v')
plt.plot(t, v2, label='axon_myelin_1(0.5).v')
plt.plot(t, v3, label='axon_no_myelin_2(0.5).v')
plt.plot(t, v4, label='axon_myelin_2(0.5).v')
plt.plot(t, v5, label='axon_no_myelin_3(0.5).v')
plt.plot(t, v6, label='axon_myelin_3(0.5).v')

plt.xlim((9, 30))         # set the "x limits" of the display
plt.xlabel('t (ms)')      # label the x axis
plt.ylabel('Membrane voltage (mV)')      # label the x axis

plt.legend()
mpld3.display()           # note that mpld3.display will create a draggable, zoomable plot
```
